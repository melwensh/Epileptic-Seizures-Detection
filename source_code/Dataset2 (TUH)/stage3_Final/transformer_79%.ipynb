```python
import os, glob, gc, warnings
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (Input, Conv1D, Conv2D, BatchNormalization, Activation,
                                     MaxPooling1D, GlobalAveragePooling1D, GlobalAveragePooling2D,
                                     Dense, Dropout, Concatenate, GaussianNoise, LeakyReLU,
                                     SpatialDropout1D, SpatialDropout2D, MultiHeadAttention, LayerNormalization)
from tensorflow.keras.regularizers import l2
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from tensorflow.keras.layers import Add

warnings.filterwarnings('ignore')

TARGET_CLASSES = ['fnsz', 'gnsz', 'cpsz']
LABEL_MAP = {lbl: idx for idx, lbl in enumerate(TARGET_CLASSES)}

DATA_DIR = '/content/local_data'
RESULTS_DIR = '/content/drive/MyDrive/gp_dataset/model_results_3classes_v1_split'
os.makedirs(RESULTS_DIR, exist_ok=True)

MODEL_OUT = f'{RESULTS_DIR}/Seizure_Type_Classifier_v1.h5'

class MultiClassFusionDataGenerator(tf.keras.utils.Sequence):
    def __init__(self, save_dir, split_name, batch_size=64, shuffle=True, **kwargs):
        super().__init__(**kwargs)

        self.save_dir = save_dir
        self.split_name = split_name
        self.batch_size = batch_size
        self.shuffle = shuffle

        self.raw_files = sorted(glob.glob(f'{save_dir}/X_raw_{self.split_name.lower()}_part*.npy'))
        self.eng_files = sorted(glob.glob(f'{save_dir}/X_eng_{self.split_name.lower()}_part*.npy'))
        self.y_files = sorted(glob.glob(f'{save_dir}/y_{self.split_name.lower()}_part*.npy'))

        if len(self.raw_files) == 0:
            available_files = os.listdir(save_dir)[:5] if os.path.exists(save_dir) else "Directory Not Found"
            raise FileNotFoundError(
                f"No files found for split '{self.split_name}' in: {save_dir}\n"
            )

        self.file_batch_counts = []
        for yf in self.y_files:
            length = len(np.load(yf, mmap_mode='r'))
            self.file_batch_counts.append(int(np.ceil(length / self.batch_size)))

        self.file_indices = np.arange(len(self.raw_files))
        self.batch_map = []
        self.on_epoch_end()

    def __len__(self):
        return int(sum(self.file_batch_counts))

    def on_epoch_end(self):
        self.batch_map = []
        if self.shuffle:
            np.random.shuffle(self.file_indices)

        for f_idx in self.file_indices:
            num_batches = self.file_batch_counts[f_idx]
            batch_order = np.arange(num_batches)
            if self.shuffle:
                np.random.shuffle(batch_order)
            for b_idx in batch_order:
                self.batch_map.append((f_idx, b_idx))

    def __getitem__(self, index):
        f_idx, b_idx = self.batch_map[index]

        X_raw = np.load(self.raw_files[f_idx], mmap_mode='r')
        X_eng = np.load(self.eng_files[f_idx], mmap_mode='r')
        y_tmp = np.load(self.y_files[f_idx], mmap_mode='r')

        start = b_idx * self.batch_size
        end = start + self.batch_size

        y_str = np.char.strip(np.char.lower(y_tmp[start:end].astype(str)))
        y_batch = np.array([LABEL_MAP[lbl] for lbl in y_str if lbl in LABEL_MAP])

        y_batch_onehot = to_categorical(y_batch, num_classes=len(TARGET_CLASSES))

        X_raw_batch = np.array(X_raw[start:end])
        X_eng_batch = np.array(X_eng[start:end])

        return (X_raw_batch, X_eng_batch), y_batch_onehot

train_gen = MultiClassFusionDataGenerator(DATA_DIR, "train", batch_size=64, shuffle=True)
val_gen = MultiClassFusionDataGenerator(DATA_DIR, "val", batch_size=64, shuffle=False)
print(f"Data generators initialized. Train batches: {len(train_gen)}, Validation batches: {len(val_gen)}")

def compute_class_weights_safe(generator, split_name):
    print(f"Computing class distribution and weights for {split_name.upper()} set...")
    class_counts = {idx: 0 for idx in range(len(TARGET_CLASSES))}
    total_samples = 0

    for yf in generator.y_files:
        y_tmp = np.load(yf, mmap_mode='r')
        chunk_size = 50000
        for i in range(0, len(y_tmp), chunk_size):
            chunk = y_tmp[i:i+chunk_size]
            chunk_str = np.char.strip(np.char.lower(chunk.astype(str)))

            y_indices = [LABEL_MAP[lbl] for lbl in chunk_str if lbl in LABEL_MAP]
            for idx in y_indices:
                class_counts[idx] += 1
                total_samples += 1

            del chunk, chunk_str, y_indices
        del y_tmp
        gc.collect()

    if total_samples == 0:
        raise ValueError("Zero valid samples detected.")

    for idx, count in class_counts.items():
        percentage = (count / total_samples) * 100
        print(f"{TARGET_CLASSES[idx].upper()}: {count} samples ({percentage:.2f}%)")
    print(f"Total samples: {total_samples}")

    class_weights = {}
    for idx, count in class_counts.items():
        if count > 0:
            class_weights[idx] = total_samples / (len(TARGET_CLASSES) * count)
        else:
            class_weights[idx] = 1.0

    print(f"Computed class weights: {class_weights}")
    return class_weights

train_class_weights = compute_class_weights_safe(train_gen, "Train")
_ = compute_class_weights_safe(val_gen, "Validation")

def build_strong_multiclass_model(raw_shape=(1280, 20), eng_shape=(5, 5, 5), num_classes=3):
    reg = l2(0.002)

    input_raw = Input(shape=raw_shape, name="Raw_Input")
    r = GaussianNoise(0.15)(input_raw)

    r = Conv1D(64, kernel_size=15, strides=2, padding='same', kernel_regularizer=reg)(r)
    r = LeakyReLU(alpha=0.1)(r)
    r = MaxPooling1D(pool_size=4)(r)
    r = Dropout(0.3)(r)

    attn_output = MultiHeadAttention(num_heads=4, key_dim=64, kernel_regularizer=reg)(r, r)
    attn_output = Dropout(0.4)(attn_output)
    r_attn = Add()([r, attn_output])
    r_attn = LayerNormalization()(r_attn)

    ff_output = Dense(64, kernel_regularizer=reg)(r_attn)
    ff_output = LeakyReLU(alpha=0.1)(ff_output)
    ff_output = Dropout(0.3)(ff_output)
    r_ff = Add()([r_attn, ff_output])
    r1 = LayerNormalization()(r_ff)

    r3 = GlobalAveragePooling1D()(r1)

    input_eng = Input(shape=eng_shape, name="Engineered_Input")
    e = Conv2D(8, (3, 3), padding='same', kernel_regularizer=reg)(input_eng)
    e = LeakyReLU(alpha=0.1)(e)
    e = Dropout(0.1)(e)

    e = Conv2D(16, (3, 3), padding='same', kernel_regularizer=reg)(e)
    e = LeakyReLU(alpha=0.1)(e)
    e = GlobalAveragePooling2D()(e)

    merged = Concatenate()([r3, e])

    dense1 = Dense(32, kernel_regularizer=reg)(merged)
    dense1 = LeakyReLU(alpha=0.1)(dense1)
    dense1 = Dropout(0.6)(dense1)

    dense2 = Dense(64, kernel_regularizer=reg)(dense1)
    dense2 = LeakyReLU(alpha=0.1)(dense2)
    dense2 = Dropout(0.5)(dense2)

    outputs = Dense(num_classes, activation='softmax', name="Output_Classification")(dense2)

    return Model(inputs=[input_raw, input_eng], outputs=outputs)

model = build_strong_multiclass_model(num_classes=len(TARGET_CLASSES))
optimizer = tf.keras.optimizers.Adam(learning_rate=0.0001)

model.compile(
    optimizer=optimizer,
    loss=tf.keras.losses.CategoricalCrossentropy(label_smoothing=0.15),
    metrics=['accuracy']
)

callbacks = [
    EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True, mode='min', verbose=1),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=4, min_lr=1e-6, mode='min', verbose=1),
    ModelCheckpoint(MODEL_OUT, monitor='val_loss', save_best_only=True, mode='min', verbose=1)
]

history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=60,
    callbacks=callbacks,
    class_weight=train_class_weights,
    verbose=1
)

test_gen = MultiClassFusionDataGenerator(DATA_DIR, "test", batch_size=64, shuffle=False)

probs_test = model.predict(test_gen)
preds_test = np.argmax(probs_test, axis=1)

y_test_true = []
for yf in test_gen.y_files:
    y_tmp = np.load(yf, mmap_mode='r')
    chunk_size = 50000
    for i in range(0, len(y_tmp), chunk_size):
        chunk = y_tmp[i:i+chunk_size]
        y_str = np.char.strip(np.char.lower(chunk.astype(str)))
        y_test_true.extend([LABEL_MAP[lbl] for lbl in y_str if lbl in LABEL_MAP])
        del chunk, y_str
    del y_tmp
    gc.collect()

y_test_true = np.array(y_test_true)

print(classification_report(y_test_true, preds_test, target_names=TARGET_CLASSES))

plt.figure(figsize=(8, 6))
cm = confusion_matrix(y_test_true, preds_test)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=TARGET_CLASSES, yticklabels=TARGET_CLASSES)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```
